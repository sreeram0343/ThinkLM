```python
from datasets import load_dataset

dataset = load_dataset("rajpurkar/squad", split="validation")

print(len(dataset))
print(dataset[0]["question"])
```

```python
import json
import os

anchor=[]

for sample in dataset.select(range(100)):

    anchor.append({

        "id":sample["id"],

        "question":sample["question"],

        "context":sample["context"],

        "answers":sample["answers"]["text"]

    })

os.makedirs("data",exist_ok=True)

with open("data/squad_anchor.json","w") as f:

    json.dump(anchor,f,indent=4)

print("Saved",len(anchor),"samples")
```
